```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
```

```python
trip_fare_path = '/Users/earl/taxi/data/trip_fare_4.csv'
trip_data_path = '/Users/earl/taxi/data/trip_data_4.csv'
```

```python
trip_fare = pd.read_csv(trip_fare_path)
trip_fare.columns
```

```python
trip_fare
```

```python
trip_data = pd.read_csv(trip_data_path)
trip_data.columns
```

```python
trip_data.sort_values(' pickup_datetime')
```

```python
trip_data
```

```python
# BQ a.
passco = trip_data[[' passenger_count']]
```

```python
passco_freq = passco.groupby(' passenger_count')[' passenger_count'].count().to_frame(name='frequency')
```

```python
passco_freq
```

```python
passco_nzero = passco[passco[' passenger_count'] != 0]
```

```python
passco_nzero_freq = passco_nzero.groupby(' passenger_count')[' passenger_count'].count().to_frame(name='frequency')
```

```python
passco_nzero_freq
```

```python
sns.countplot(x=' passenger_count', data=passco_nzero)

ax_pos = -0.05

plt.xlabel('Number of passenger', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(ax_pos, 1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```

```python
# BQ b.
paytype = trip_fare[[' payment_type']]
paytype_nzero = paytype[passco[' passenger_count'] != 0]
paytype_nzero_freq = paytype_nzero.groupby(' payment_type')[' payment_type'].count().to_frame(name='frequency')
paytype_nzero_freq
```

```python
order = sorted(paytype_nzero[' payment_type'].value_counts().index)
sns.countplot(x=' payment_type', data=paytype_nzero, order = order)

ax_pos = -0.05

plt.xlabel('Payment type', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(-0.05,1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```

```python
# BQ c.
fareamo = trip_fare[[' fare_amount']]
fareamo_nzero = fareamo[passco[' passenger_count'] != 0]
bins = np.arange(0.5, 51.5, 1)
fareamo_nzero['fare_amount_binned'] = pd.cut(fareamo_nzero[' fare_amount'], bins)
fareamo_nzero_freq = fareamo_nzero.groupby('fare_amount_binned')['fare_amount_binned'].count().to_frame(name='frequency')
fareamo_nzero_freq
```

```python
bins = np.arange(0, 52)
print(bins)

sns.distplot(fareamo_nzero[' fare_amount'], hist=True, kde=False, bins=bins-0.5, hist_kws={'ec':"k"})

ax_pos = -0.05

plt.xticks(np.arange(bins[0], bins[-1], 5))
plt.xlim([bins[0], bins[-1]])
plt.xlabel('Fare amount ($)', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(ax_pos,1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```

```python
# BQ d.
tipamo = trip_fare[[' tip_amount']]
tipamo_nzero = tipamo[passco[' passenger_count'] != 0]
bins = np.arange(0, 15.5, 0.5)
tipamo_nzero['tip_amount_binned'] = pd.cut(tipamo_nzero[' tip_amount'], bins)
tipamo_nzero_freq = tipamo_nzero.groupby('tip_amount_binned')['tip_amount_binned'].count().to_frame(name='frequency')
tipamo_nzero_freq
```

```python
bins = np.arange(0, 52)
print(bins)

sns.distplot(fareamo_nzero[' fare_amount'], hist=True, kde=False, bins=bins-0.5, hist_kws={'ec':"k"})

ax_pos = -0.05

plt.xticks(np.arange(bins[0], bins[-1], 5))
plt.xlim([bins[0], bins[-1]])
plt.xlabel('Fare amount ($)', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(ax_pos,1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))

# bins_plot = bins+0.5
# print(bins_plot)
plt.bar(bins[:-1], tipamo_nzero_freq.frequency.values, align='edge', width=bins[1]-bins[0], edgecolor='k')
plt.grid(linewidth=0.5)
plt.xticks(np.arange(bins[0],bins[-1],1))
plt.xlim([bins[0],bins[-1]])
plt.ylabel('Trip count')
plt.xlabel('Tip amount ($)')
plt.title('Distribution of tip amounts')
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```
