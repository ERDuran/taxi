```python
trip_fare_path = '/Users/earl/taxi/data/trip_fare_4.csv'
trip_data_path = '/Users/earl/taxi/data/trip_data_4.csv'
```

```python
trip_fare = pd.read_csv(trip_fare_path)
trip_fare.columns
```

```python
trip_fare
```

```python
trip_data = pd.read_csv(trip_data_path)
trip_data.columns
```

```python
trip_data.sort_values(' pickup_datetime')
```

```python
trip_data
```

```python
# BQ a.
passco = trip_data[[' passenger_count']]
```

```python
passco_freq = passco.groupby(' passenger_count')[' passenger_count'].count().to_frame(name='frequency')
```

```python
passco_freq
```

```python
passco_nzero = passco[passco[' passenger_count'] != 0]
```

```python
passco_nzero_freq = passco_nzero.groupby(' passenger_count')[' passenger_count'].count().to_frame(name='frequency')
```

```python
passco_nzero_freq
```

```python
plt.bar(passco_nzero_freq.index, passco_nzero_freq.frequency)
plt.grid(linewidth=0.5)
plt.xticks(range(10))
plt.ylabel('Trip count')
plt.xlabel('Number of passenger')
plt.title('Distribution of number of passengers per trip')
# plt.gca().ticklabel_format(style='plain')
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```

```python
# BQ b.
paytype = trip_fare[[' payment_type']]
paytype_nzero = paytype[passco[' passenger_count'] != 0]
paytype_nzero_freq = paytype_nzero.groupby(' payment_type')[' payment_type'].count().to_frame(name='frequency')
paytype_nzero_freq
```

```python
plt.bar(paytype_nzero_freq.index, paytype_nzero_freq.frequency)
plt.grid(linewidth=0.5)
# plt.xticks(range(10))
plt.ylabel('Trip count')
plt.xlabel('Payment type')
plt.title('Distribution of payment types')
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```

```python
# BQ c.
fareamo = trip_fare[[' fare_amount']]
fareamo_nzero = fareamo[passco[' passenger_count'] != 0]
bins = np.arange(0, 31, 1)
fareamo_nzero['fare_amount_binned'] = pd.cut(fareamo_nzero[' fare_amount'], bins)
fareamo_nzero_freq = fareamo_nzero.groupby('fare_amount_binned')['fare_amount_binned'].count().to_frame(name='frequency')
fareamo_nzero_freq
```

```python
bins_plot = bins+0.5
# print(bins_plot)
plt.bar(bins_plot[:-1], fareamo_nzero_freq.frequency.values, width=1.0)
plt.grid(linewidth=0.5)
plt.xticks(np.arange(bins[0],bins[-1],2))
plt.xlim([bins[0],bins[-1]])
plt.ylabel('Trip count')
plt.xlabel('Fare amount')
plt.title('Distribution of fare amounts')
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```

```python
# BQ d.
tipamo = trip_fare[[' tip_amount']]
tipamo_nzero = tipamo[passco[' passenger_count'] != 0]
bins = np.arange(0, 15.5, 0.5)
tipamo_nzero['tip_amount_binned'] = pd.cut(tipamo_nzero[' tip_amount'], bins)
tipamo_nzero_freq = tipamo_nzero.groupby('tip_amount_binned')['tip_amount_binned'].count().to_frame(name='frequency')
tipamo_nzero_freq
```

```python
bins_plot = bins+0.5
# print(bins_plot)
plt.bar(bins_plot[:-1], tipamo_nzero_freq.frequency.values, width=1.0)
plt.grid(linewidth=0.5)
plt.xticks(np.arange(bins[0],bins[-1],1))
plt.xlim([bins[0],bins[-1]])
plt.ylabel('Trip count')
plt.xlabel('Tip amount')
plt.title('Distribution of tip amounts')
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```
