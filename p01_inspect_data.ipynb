```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
def arange(start,end,stride=1):
    return np.arange(start,end+0.00000001,stride)
```

```python
trip_fare_path = '/Users/earl/taxi/data/trip_fare_4.csv'
trip_data_path = '/Users/earl/taxi/data/trip_data_4.csv'
```

```python
trip_fare = pd.read_csv(trip_fare_path)
trip_fare.columns
```

```python
trip_fare
```

```python
trip_data = pd.read_csv(trip_data_path)
trip_data.columns
```

```python
trip_data
```

```python
# BQ a.
passco_raw = trip_data[[' passenger_count']]
passco_freq = passco_raw.groupby(' passenger_count')[' passenger_count'].count().to_frame(name='frequency')
passco_freq
```

```python
trip_fare_nzero = trip_fare[passco_raw[' passenger_count'] != 0]
trip_data_nzero = trip_data[passco_raw[' passenger_count'] != 0]
```

```python
passco = trip_data_nzero[[' passenger_count']]
passco_freq = passco.groupby(' passenger_count')[' passenger_count'].count().to_frame(name='frequency')
passco_freq
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
```

```python
sns.countplot(x=' passenger_count', data=passco)

ax_pos = -0.05

plt.xlabel('Number of passenger', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(ax_pos, 1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```

```python
# BQ b.
paytype = trip_fare_nzero[[' payment_type']]
paytype_freq = paytype.groupby(' payment_type')[' payment_type'].count().to_frame(name='frequency')
paytype_freq
```

```python
order = sorted(paytype[' payment_type'].value_counts().index)
sns.countplot(x=' payment_type', data=paytype, order=order)

ax_pos = -0.05

plt.xlabel('Payment type', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(-0.05,1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```

```python
# BQ c.
fareamo = trip_fare_nzero[[' fare_amount']]
bins = arange(0,32)-0.5
fareamo['fare_amount_binned'] = pd.cut(fareamo[' fare_amount'], bins=bins, right=False)
fareamo_freq = fareamo.groupby('fare_amount_binned')['fare_amount_binned'].count().to_frame(name='frequency')
fareamo_freq
```

```python
sns.distplot(fareamo[' fare_amount'], hist=True, kde=False, bins=bins, hist_kws={'ec':"k"})

ax_pos = -0.05

bins_label = bins + 0.5
plt.xticks(bins_label[::2])
plt.xlim([bins_label[0], bins[-2]])
plt.xlabel('Fare amount ($)', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(ax_pos,1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```

```python
# BQ f.
putime = trip_fare_nzero[[' pickup_datetime']]
puhour = pd.DataFrame({'hour_24': [f[11:13] for f in putime[' pickup_datetime']]})
puhour_freq = puhour.groupby('hour_24')['hour_24'].count().to_frame(name='frequency')
puhour_freq

puhour_freq
```

```python
sns.countplot(x='hour_24', data=puhour)
plt.gcf().set_figwidth(plt.gcf().get_figwidth()*2)
ax_pos = -0.025

plt.xlabel('Pick up time (hour of the day)', fontsize=15, fontweight='black', color = '#333F4B')

plt.ylabel('Trip count', fontsize=15, fontweight='black', color = '#333F4B', rotation='horizontal', ha='right')
plt.gca().yaxis.set_label_coords(ax_pos, 1.02)
plt.gca().yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plt.gca().get_yaxis().set_ticks(arange(0,800000,200000))

plt.gca().tick_params(axis='both', which='major', labelsize=12)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['left'].set_smart_bounds(True)
plt.gca().spines['bottom'].set_smart_bounds(True)
plt.gca().spines['left'].set_position(('axes', ax_pos))
```
